```python
import seaborn as sns
import matplotlib.pyplot as plt
import tensorboard as tb
```

```python
def filter_df(df, tag: str):
    """
    tag: epoch_accuracy | epoch_loss
    """
    df_validation = df[df.run.str.endswith("/validation")]
    runs = df_validation.run.unique()
    df_filtered = df_validation[df_validation['tag']==tag].pivot("step", "run", "value")
    return df_filtered

def plot(df, tag, plot_title):
    sns.lineplot(data=filter_df(df, tag)).set(title=plot_title)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

# Keypoint analysis

```python
experiment_keypoint = tb.data.experimental.ExperimentFromDev("Jv0WvrspRamhpLQngYNwUw")
df_keypoint = experiment_keypoint.get_scalars()
```

## Accuracy

```python
plot(df_keypoint, "epoch_accuracy", "Accuracy")
```

## Loss

```python
plot(df_keypoint, "epoch_loss", "Loss")
```
